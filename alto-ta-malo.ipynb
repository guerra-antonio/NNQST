```python
from classical_shadow import sample_measurements, random_state
import matplotlib.pyplot as plt
import torch
```

```python
psi = random_state()
psi
```

```python
psi = torch.tensor([[0.5475 + 0j],
                   [0.3546 + 0.758j]])
psi
```

```python
data = sample_measurements(psi = psi, error = 0.2, M=25)
```

```python
X = []
Y = []
Z = []

for _ in data:
    bit, basis = _

    if basis == "X":
        X.append(bit.item())
    elif basis == "Y":
        Y.append(bit.item())
    else:
        Z.append(bit.item())
```

```python
print(len(X), len(Y), len(Z))
```

```python
basis = {
    "sigmaX" : X,
    "sigmaY" : Y,
    "sigmaZ" : Z,
}
```

```python
def make_hist(basis_label):
    plt.hist(basis[basis_label], bins = [0.0, 0.5, 1.0], rwidth = 0.8)
    plt.grid()
    plt.title(basis_label)
```

```python
make_hist("sigmaX")
```

```python
make_hist("sigmaY")
```

```python
make_hist("sigmaZ")
```
